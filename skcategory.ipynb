```python
lemma = WordNetLemmatizer()
```

```python
pol = pd.read_csv('data/political_tweets.csv')
pol = pol['text']
pol = pol.drop_duplicates()
len(pol)
```

```python
ent = pd.read_csv('data/entertainment_tweets.csv')
ent = ent['text']
ent = ent.drop_duplicates()
len(ent)
```

```python
other = pd.read_csv('data/other_tweets.csv')
other = other['text']
other = other.drop_duplicates()
len(other)
```

```python
tech = pd.read_csv('data/technology_tweets.csv')
tech = tech['text']
tech = tech.drop_duplicates()
len(tech)
```

```python
ids = pd.read_csv('data/Followers.csv')
```

```python
x = ids['tech'].dropna()
y = ids['sports'].dropna()
z = ids['entertainment'].dropna()
w = ids['politics'].dropna()
```

```python
tech = pd.concat([tech, x])
spo = pd.concat([spo, y])
ent = pd.concat([ent, z])
pol = pd.concat([pol, w])
```

```python
data = pd.concat([pol,ent,env,spo,other,tech])
```

```python
type(data)
```

```python
labels = []
for t in range(len(pol)):
    labels.append('pol')
for t in range(len(ent)):
    labels.append('ent')
for t in range(len(env)):
    labels.append('env')
for t in range(len(spo)):
    labels.append('spo')
for t in range(len(other)):
    labels.append('other')
for t in range(len(tech)):
    labels.append('tech')
```

```python
tweets_tokenized = []
for tweet in data:
    tweets_tokenized.append(nltk.wordpunct_tokenize(tweet))
```

```python
pattern = re.compile('^http|^:|^[\']|^["]|^@|^[\\]|^[/]|^x[a-z0-9]+|^[0-9]+|^[?]|^[.]|^[#]|^[!]|^[\'@]|^[\"@]|^[,]|^[:]|^[;]')
```

```python
temp = copy.deepcopy(tweets_tokenized)
for i in range(0,len(temp)):
    for j in temp[i]:
        if pattern.match(j) or len(j)<3:
            tweets_tokenized[i].remove(j)
```

```python
tokenized_tweets = []
for t in tweets_tokenized:
    tokenized_tweets.append(' '.join(t))
```

```python
tokenized_tweets
```

```python
lemmatized_tweets = []
for t in tokenized_tweets:
    str = ""
    for k in t.split():
        str += lemma.lemmatize(k)+" "
    lemmatized_tweets.append(str)
```

```python
lemmatized_tweets[49000:]
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(lemmatized_tweets, labels, test_size=0.1, random_state=50)
```

```python
vectorizer = TfidfVectorizer(sublinear_tf = True, stop_words='english', lowercase = True)
```

```python
features_train_transformed = vectorizer.fit_transform(features_train)
features_test_transformed  = vectorizer.transform(features_test)
```

```python
file = open('vectorizer.pickle','wb')
pickle.dump(vectorizer, file)
file.close()
```
